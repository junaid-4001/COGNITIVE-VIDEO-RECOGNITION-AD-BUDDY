```python
from nltk.tokenize import word_tokenize

text = letters
tokens = word_tokenize(text)
```

```python
from nltk.corpus import stopwords
import string

stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)

tokens = [token.lower() for token in tokens]
tokens = [token for token in tokens if token not in stop_words and token not in punctuation]
```

```python
from collections import defaultdict

word_freq = defaultdict(int)
word_degree = defaultdict(int)

for token in tokens:
    word_freq[token] += 1
    for word in tokens:
        if word != token:
            word_degree[token] += 1

word_scores = {token: (word_degree[token] / word_freq[token]) for token in tokens}
```

```python
num_keywords = 100
ranked_keywords = sorted(word_scores, key=word_scores.get, reverse=True)[:num_keywords]
print(ranked_keywords)
```

```python
from nltk.corpus import wordnet

def are_synonyms(word1, word2):
    synsets1 = wordnet.synsets(word1)
    synsets2 = wordnet.synsets(word2)

    for synset1 in synsets1:
        for synset2 in synsets2:
            if synset1.wup_similarity(synset2) is not None and synset1.wup_similarity(synset2) > 0.7:
                return True

    return False

my_dict = {}

for key, value in names.items():
    for word in ranked_keywords:
        if are_synonyms(value, word):
            if value in my_dict.keys():
                my_dict[value] += 1 
            else:
                my_dict[value] = 1

            print(f"{value} and {word} are synonyms. ({key} matches)")
            

print(my_dict)
```
